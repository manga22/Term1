```
def project(A, b, u):
  A_inv = np.matmul(A.T,np.linalg.inv(np.matmul(A,A.T)))
  x= u - np.matmul(A_inv,(np.matmul(A,u) - b))
  return x
```

```
def projected_grad_descent(X0, C, d, A, b, step=0.5, eps=0.001):
  grad = lambda x: np.matmul(C.T,np.matmul(C,x)+d)
  proj = lambda x: project(A,b,x)
  Q_inv = np.linalg.inv(np.matmul(C.T,C))
  Z =np.matmul(Q_inv, np.matmul(A.T,np.linalg.inv(np.matmul(A,np.matmul(Q_inv, A.T)))))
  x_opt = -np.matmul(Q_inv,np.matmul(C.T,d))+np.matmul(Z,b+np.matmul(np.matmul(A,np.matmul(Q_inv,C.T)),d))
  x = X0
  k=0
  #print(x_opt)
  while(np.linalg.norm(x-x_opt)>eps):
    x = proj(x-step*grad(x))
    k+=1
    #print(k, np.linalg.norm(x-x_opt), x, grad(x))
  print(f'X0 = {X0}, step = {step}, eps = {eps}, iters:{k}')
  #print(x_opt, x)
  return k
```

```
A = np.array([[1, -1, 0],[0,1,-1]])
b = np.array([2,3])
C = np.array([[math.sqrt(2),0,0],[0,1,0],[0,0,1]])
d = np.array([2,3,1])
X0 = np.array([0,0,0])
```

```
import csv
import numpy as np
data = np.genfromtxt('a4.csv', delimiter=',',dtype= float).T
y = data[:,2]
A = np.ones((10,3))
X = np.ones((10,2))
for i in range(10):
  X[i] = data[i,:2]
  A[i,:2]=data[i,:2]
  A[i]*=y[i]
print(A)
print(X.T)
```
